```python
def Sep_hist(MA):
    column_name = 'MA'+str(MA)
    
    BTC_D[column_name] = BTC_D['close'].sort_index(ascending=False).rolling(window=MA).mean().sort_index(ascending=True)
    BTC_D['index2'] = BTC_D.index
    
    CHTM = BTC_D[BTC_D[column_name] < BTC_D['close']]
    CHTM['Sep'] = (1 - CHTM['close']/CHTM[column_name])*100
    CHTM['Sep'].hist()
```

```python
Sep_hist(20)
# 이거는 그래프로도 좀 봐야겠다.
```

```python
def Sep_hist2(MA):
    column_name = 'MA'+str(MA)
    
    BTC_D[column_name] = BTC_D['close'].sort_index(ascending=False).rolling(window=MA).mean().sort_index(ascending=True)
    BTC_D['index2'] = BTC_D.index
    
    CLTM = BTC_D[BTC_D[column_name] > BTC_D['close']]
    CLTM['Sep'] = (1 - CLTM['close']/CLTM[column_name])*100
    CLTM['Sep'].hist()
```
